```python
# Studi Kasus Sederhana: Segmentasi Pelanggan Berdasarkan Lokasi Belanja

from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# 1. Buat data dummy pelanggan
data, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# 2. Fungsi label grid
def get_grid_label(point, size):
    return tuple((point // size).astype(int))

# 3. Pemetaan ke grid
grid_size = 1.0
grid_dict = {}
for point in data:
    label = get_grid_label(point, grid_size)
    grid_dict.setdefault(label, []).append(point)

# 4. Ambil sel padat (5 titik)
threshold = 5
clusters = [np.array(pts) for pts in grid_dict.values() if len(pts) >= threshold]

# 5. Visualisasi dengan label cluster
plt.figure(figsize=(7, 6))
for idx, cl in enumerate(clusters):
    plt.scatter(cl[:, 0], cl[:, 1], label=f"Cluster {idx+1}")
    # Tambahkan label tengah cluster
    center = np.mean(cl, axis=0)
    plt.text(center[0], center[1], f"C{idx+1}", fontsize=10, weight='bold', color='black')

plt.title("Segmentasi Pelanggan (Grid-Based Clustering)")
plt.xlabel("Fitur Lokasi X")
plt.ylabel("Fitur Lokasi Y")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
